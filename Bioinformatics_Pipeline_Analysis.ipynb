```python
!pip install biopython

import os
from Bio import SeqIO
import pandas as pd

# Pipeline ayarları:
# Kod, çalıştırıldığı klasörün altındaki 'sequences' klasörünü arar.
# Buraya örnek bir dosya yolu tanımlıyoruz.
SEQUENCE_DIR = "./sequences"

def process_sequences(directory):
    """
    Ham sekansları okur, uçlarını kırpar ve filtreler.
    Bu 'Pipeline', laboratuvar verilerini standartlaştırmak için yazılmıştır.
    """
    processed_data = []

    # Eğer dizin yoksa hata vermesin, boş liste dönsün
    if not os.path.exists(directory):
        print(f"Uyarı: '{directory}' klasörü bulunamadı. Lütfen örnek verilerini buraya yükle.")
        return None

    for filename in os.listdir(directory):
        if filename.endswith(".ab1"):
            file_path = os.path.join(directory, filename)

            # SeqIO ile okuma
            try:
                record = SeqIO.read(file_path, "abi")
                # 40-50 baz kırpma (Kalite kontrolü)
                trimmed_record = record[40:-50]

                # Sadece uzunluğu makul olanları al
                if len(trimmed_record.seq) > 100:
                    processed_data.append({
                        "Sample_ID": filename.replace(".ab1", ""),
                        "Sequence_Length": len(trimmed_record.seq),
                        "Status": "Passed QC"
                    })
            except Exception as e:
                print(f"Dosya işlenirken hata: {filename} -> {e}")

    return pd.DataFrame(processed_data)

# Analizi çalıştır
df_results = process_sequences(SEQUENCE_DIR)
print("Pipeline başarıyla çalıştı! İşlenen veriler:")
if df_results is not None:
    print(df_results.head())
```
