```python

!pip install biopython

"""
Central Dogma Simulator: Smart Ribosome Edition
Author: Ayten
Description: This bioinformatics pipeline automatically fetches a specified reference gene
from the NCBI database using an Accession ID, simulates RNA Polymerase transcription,
and uses a "Smart Ribosome" logic (finding the AUG start codon and stopping at stop codons)
to translate the mRNA into a clean protein sequence.
"""

from Bio import Entrez, SeqIO

# --- PIPELINE CONFIGURATION ---
# NCBI requires an email address to track usage and prevent server overload.
Entrez.email = "harmanayten07@gmail.com"

def run_central_dogma_pipeline(accession_id):
    """
    Fetches nucleotide data from NCBI, saves it locally as a FASTA file,
    and performs biologically accurate transcription and translation.
    """
    file_name = f"{accession_id}_sequence.fasta"

    # --- STEP 1: AUTOMATED DATA FETCHING ---
    print(f"[*] Connecting to NCBI... Fetching Accession ID: {accession_id}")
    try:
        # efetch: The NCBI agent that downloads our target data
        handle = Entrez.efetch(db="nucleotide", id=accession_id, rettype="fasta", retmode="text")
        gene_record = SeqIO.read(handle, "fasta")
        handle.close() # Always close the connection to NCBI servers

        # Save the real biological data locally
        SeqIO.write(gene_record, file_name, "fasta")
        print(f"[+] Success! Sequence saved locally as '{file_name}'\n")

    except Exception as e:
        print(f"[-] Error fetching data from NCBI: {e}")
        return

    # --- STEP 2: SMART BIOLOGICAL SIMULATION (CENTRAL DOGMA) ---
    print("-" * 50)
    print("🔬 SMART RIBOSOME SIMULATION RESULTS")
    print("-" * 50)

    dna_sequence = gene_record.seq

    # 1. Transcription (DNA -> mRNA)
    mrna_sequence = dna_sequence.transcribe()

    # 2. Smart Translation (mRNA -> Protein)
    # Searching for the Start Codon (AUG / Methionine)
    start_index = mrna_sequence.find("AUG")

    if start_index != -1:
        print(f"[INFO] Ribosome identified the START codon (AUG) at nucleotide {start_index}.")

        # Trimming the 5' UTR to establish the correct reading frame
        coding_mrna = mrna_sequence[start_index:]

        # to_stop=True ensures translation halts exactly at the first STOP codon
        protein_sequence = coding_mrna.translate(to_stop=True)

    else:
        print("[WARNING] No START codon (AUG) found in this sequence!")
        protein_sequence = "Translation Failed."

    # --- STEP 3: DISPLAYING RESULTS ---
    # Displaying truncated strings for a cleaner console output
    print(f"Record Info   : {gene_record.description[:70]}...")
    print(f"DNA (5'->3')  : {dna_sequence[:60]}...")
    print(f"Full mRNA     : {mrna_sequence[:60]}...")
    print(f"Clean Protein : {protein_sequence[:60]}...")
    print(f"Total Protein Length: {len(protein_sequence)} amino acids")
    print("-" * 50)
    print("Pipeline Execution Completed.")

# Run the pipeline with the Human TP53 Tumor Suppressor Gene Reference
if __name__ == "__main__":
    TARGET_GENE_ID = "NM_000546"
    run_central_dogma_pipeline(TARGET_GENE_ID)

```
